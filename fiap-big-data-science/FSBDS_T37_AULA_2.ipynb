```python
#!pip install pandas
import pandas as pd

df = pd.read_csv("diamonds.csv") 
```

```python
df.head()
```

```python
df.shape
```

```python
a1 = df.sample(n=3000)

media_df = df[['price']].mean()
media_a1 = a1[['price']].mean()
diff = ((media_df.values - media_a1.values) / media_df.values) * 100

print('Média Original: %.2f - Média Amostra: %.2f - Diff %.2f%%' % (media_df.values, media_a1.values, diff))
a1.head()
```

```python
df_a = pd.concat([df[['price']].describe(), a1[['price']].describe()], axis=1)
df_a.columns = ['price_orign', 'price_sample']
df_a['diff'] = ((df_a['price_orign'] - df_a['price_sample']) / df_a['price_orign']) * 100
df_a.round(2)
```
